```python
import requests as re
from bs4 import BeautifulSoup as bs
import pandas as pd



book_name = []
book_price = []
book_link = []
book_rating =[]


for i in range(1,51):
  resp = re.get(f"https://books.toscrape.com/catalogue/page-{i}.html")
  soup = bs(resp.text,"html.parser")
  name = soup.find_all("article",class_="product_pod")
  price = soup.find_all("div",class_="product_price")
  link = soup.find_all("article",class_="product_pod")
  rating = soup.find_all("article",class_="product_pod")

  for k in range(20):
    book_name.append(name[k].find("a",title = True).attrs["title"])
    book_price.append(price[k].find("p").text.strip("A"))
    links = link[k].find("a").attrs
    book_link.append("https://books.toscrape.com/catalogue/")
    books = rating[k].find("p").attrs
    book_rating.append(books["class"][1])
```

```python
df = pd.DataFrame(list(zip(book_name,book_price,book_link,book_rating)),columns = ["Name","Price","Link","Rating"])
display(df)
```
