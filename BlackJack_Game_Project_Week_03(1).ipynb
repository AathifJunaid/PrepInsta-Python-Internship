```python
import random as rn



def deal_card():
  cards = [11,2,3,4,5,6,7,8,9,10,10,10,10]
  return rn.choice(cards)



def calculate_score(card):
  if(sum(card)==21 and len(card)==2):
    return 0

  if(sum(card)>21 and 11 in card):
    card.remove(11)
    card.append(1)

  return sum(card)


def compare(user,comp):
  if(user>21 and comp>21):
    return("You Loose")
  elif(user == comp):
    return ("Draw")
  elif(comp == 0):
    return ("Computer wins with a black jack, You Loose")
  elif(user == 0):
    return ("You win with a black jack ")
  elif(user > 21):
    return ("You loose")
  elif(comp > 21):
    return ("You win")
  elif (user > comp):
    return ("You win")
  elif (comp > user):
     ("Computer wins, You Loose")

def play_game():

  user_cards = []
  comp_cards = []

  for i in range(2):
    user_cards.append(deal_card())
    comp_cards.append(deal_card())

  while True:

    user_score = calculate_score(user_cards)
    comp_score = calculate_score(comp_cards)

    print(f"Your cards:{user_cards} , Score:{user_score}")
    print(f"The first card of the computer is:[{comp_cards[0]}]")

    if(user_score==0 or comp_score==0 or user_score > 21):
      print(f"Computer cards:[{comp_cards}] , Score:{comp_score}")
      break
    else:
      user_deal = input("Type 'Y' to get another card, else type 'N' to pass:").upper()
      print()

      if (user_deal == "Y"):
        user_cards.append(deal_card())
        user_score = calculate_score(user_cards)

      elif(user_deal == "N"):
        break

      else:
        print("Incorrect Input")
        break


  while(comp_score!=0 and comp_score<17):
    comp_cards.append(deal_card())
    comp_score = calculate_score(comp_cards)



  print(f"Your cards:{user_cards} , Score:{user_score}")
  print(f"Computer cards:{comp_cards} , Score:{comp_score}")

  print(compare(user_score,comp_score))





while True:
  print("Do you want to start the game?")
  "If Yes type 'Y', else type'N' "
  user_decision = input("If Yes type 'Y', else type'N':").upper()
  print()

  if(user_decision == "Y"):
    play_game()

  else:
    break









```
